```python
# it shoul be more generic loader of arff files, but CKD dataset has problems so here is preprocessing:
# piece of code from https://stackoverflow.com/questions/62653514/open-an-arff-file-with-scipy-io
import pandas as pd

data = []
with open('.\dataset\chronic_kidney_disease_full.arff', "r") as f:
    for line in f:
        line = line.replace('\n', '')
        data.append(line.split(','))


names = ['age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba',
         'bgr', 'bu',  'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc',
         'rbcc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane',
         'class', 'no_name']

# actual data starts from 144 line for this arff file
df = pd.DataFrame(data[145:], columns=names)
df.drop(columns=['no_name'], inplace=True)
```

```python
df.head()
```

```python
import numpy as np

df.replace('?', np.nan, inplace=True)
```

```python
df['class'].replace(["ckd\t", "no"],["ckd", "notckd"],inplace=True)

df['dm'].replace(["\tno","\tyes"," yes"],["no","yes","yes"],inplace=True)

df['cad'].replace(["\tno"],["no"],inplace=True)
```

```python
df['rbcc'].replace("\t?",None, inplace=True)

df['wbcc'].replace("\t?",None, inplace=True)

df['pcv'].replace(["\t?","\t43"],None, inplace=True)
```
